```python
import pandas as pd

# Hitung Q1, Q3, dan IQR
Q1 = data_Superstore['Sales'].quantile(0.25)
Q3 = data_Superstore['Sales'].quantile(0.75)
IQR = Q3 - Q1

# Tentukan batas atas dan batas bawah
batas_atas = Q3 + 1.5 * IQR
batas_bawah = Q1 - 1.5 * IQR

# Identifikasi nilai ekstrem
nilai_ekstrem = data_Superstore[(data_Superstore['Sales'] < batas_bawah) | (data_Superstore['Sales'] > batas_atas)]

# Tampilkan nilai ekstrem
print(nilai_ekstrem)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Hitung matriks korelasi
correlation_matrix = data_Superstore.corr(numeric_only=True)

# Atur ukuran figure
plt.figure(figsize=(12, 10))

# Buat heatmap dengan beberapa peningkatan:
# - annot=True: Menampilkan nilai korelasi di setiap sel.
# - fmt=".2f": Memformat nilai korelasi menjadi 2 angka desimal.
# - cmap="coolwarm": Menggunakan colormap "coolwarm" untuk visualisasi yang lebih baik.
# - linewidths=0.5: Menambahkan garis pemisah antar sel.
# - linecolor="white": Mengatur warna garis pemisah menjadi putih.
# - annot_kws={"size": 10}: Mengatur ukuran font anotasi.
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm",
            linewidths=0.5, linecolor="white", annot_kws={"size": 10})

# Atur judul heatmap
plt.title("Heatmap Korelasi", fontsize=16)

# Rotasi label sumbu x dan y agar lebih mudah dibaca
plt.xticks(rotation=45, ha="right", fontsize=12)
plt.yticks(rotation=0, fontsize=12)

# Tampilkan heatmap
plt.show()
```

```python
# Menggunakan Pie Chart Untuk menggambarkan proporsi kategori dalam total, seperti pangsa pasar kategori produk
import matplotlib.pyplot as plt
import seaborn as sns

# Hitung proporsi kategori
proporsi_kategori = data_Superstore['Category'].value_counts(normalize=True)

# Buat pie chart
plt.figure(figsize=(8, 8))
plt.pie(proporsi_kategori, labels=proporsi_kategori.index, autopct='%1.1f%%', startangle=90)
plt.title('Proporsi Kategori Produk')
plt.show()
```

```python
# Tampilkan dengan Line Plot Untuk menunjukkan tren dari waktu ke waktu, seperti penjualan bulanan
import matplotlib.pyplot as plt
import seaborn as sns

data_SuperStore['Order_Date'] = pd.to_datetime(data_SuperStore['Order_Date'])
monthly_sales = data_SuperStore.groupby(data_SuperStore['Order_Date'].dt.to_period('M'))['Sales'].sum()

plt.plot(monthly_sales.index.astype(str), monthly_sales.values, marker='o')
plt.title('Tren Penjualan Bulanan')
plt.xlabel('Bulan')
plt.ylabel('Total Penjualan')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ubah tipe data 'Order_Date' menjadi datetime
data_SuperStore['Order_Date'] = pd.to_datetime(data_SuperStore['Order_Date'])

# Kelompokkan data berdasarkan bulan dan hitung total penjualan
monthly_sales = data_SuperStore.groupby(data_SuperStore['Order_Date'].dt.to_period('M'))['Sales'].sum()

# Buat line plot dengan seaborn dan matplotlib untuk tampilan yang lebih menarik
plt.figure(figsize=(12, 6))  # Atur ukuran figure
sns.lineplot(x=monthly_sales.index.astype(str), y=monthly_sales.values, marker='o', color='skyblue', linewidth=2)
plt.title('Tren Penjualan Bulanan', fontsize=16)  # Perbesar ukuran judul
plt.xlabel('Bulan', fontsize=12)  # Perbesar ukuran label x-axis
plt.ylabel('Total Penjualan', fontsize=12)  # Perbesar ukuran label y-axis
plt.xticks(rotation=45, ha='right')  # Rotasi label sumbu x agar mudah dibaca
plt.grid(True, linestyle='--', alpha=0.5)  # Tambahkan gridlines
plt.tight_layout()  # Sesuaikan layout agar tidak terpotong
plt.show()
```
