```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'TAHUN': [2019, 2020, 2021, 2022, 2023],
        'SISA PERKARA TAHUN LALU': [3, 0, 0, 0, 2],
        'PERKARA MASUK': [490, 500, 520, 670, 628],
        'PERKARA PUTUS ≤5 BLN': [491, 499, 520, 670, 626],
        'PERKARA PUTUS ≥ 5 BLN': [2, 1, 0, 0, 0],
        'PERKARA PUTUS': [493, 500, 520, 670, 626],
        'SISA PERKARA': [0, 0, 0, 2, 4],
        'JUMLAH HAKIM': [4, 5, 5, 4, 3]}
df = pd.DataFrame(data)

# Data untuk box plot
data_to_plot = [df['PERKARA MASUK'], df['PERKARA PUTUS']]

# Membuat Box Plot
plt.figure(figsize=(10, 6))
plt.boxplot(data_to_plot, labels=['Perkara Masuk', 'Perkara Putus'], patch_artist=True,
            boxprops=dict(facecolor='lightblue', color='blue'),
            whiskerprops=dict(color='blue'),
            capprops=dict(color='blue'),
            medianprops=dict(color='red'))
plt.title('Box Plot Perkara Masuk dan Perkara Putus')
plt.ylabel('Jumlah Perkara')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = {'TAHUN': [2019, 2020, 2021, 2022, 2023],
        'SISA PERKARA TAHUN LALU': [3, 0, 0, 0, 2],
        'PERKARA MASUK': [490, 500, 520, 670, 628],
        'PERKARA PUTUS ≤5 BLN': [491, 499, 520, 670, 626],
        'PERKARA PUTUS ≥ 5 BLN': [2, 1, 0, 0, 0],
        'PERKARA PUTUS': [493, 500, 520, 670, 626],
        'SISA PERKARA': [0, 0, 0, 2, 4],
        'JUMLAH HAKIM': [4, 5, 5, 4, 3]}
df = pd.DataFrame(data)

# Lebar setiap batang
width = 0.25

# Posisi batang pada sumbu x
x = np.arange(len(df['TAHUN']))

# Membuat grafik batang
fig, ax = plt.subplots(figsize=(12, 6))
rects1 = ax.bar(x - width, df['PERKARA MASUK'], width, label='Perkara Masuk', color='skyblue')
rects2 = ax.bar(x, df['PERKARA PUTUS'], width, label='Perkara Putus', color='lightcoral')
rects3 = ax.bar(x + width, df['SISA PERKARA'], width, label='Sisa Perkara', color='lightgreen')

# Memberikan label, judul, dan ticks
ax.set_ylabel('Jumlah Perkara')
ax.set_title('Jumlah Perkara Masuk, Putus, dan Sisa Perkara per Tahun')
ax.set_xticks(x)
ax.set_xticklabels(df['TAHUN'])
ax.legend()

# Menambahkan label nilai di atas setiap batang
def autolabel(rects):
    """Menambahkan label nilai di atas setiap batang."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

# Menambahkan keterangan pada grafik
plt.text(0.05, 0.95, 'Sumber Data: Data Pengadilan', transform=ax.transAxes, fontsize=10,
         verticalalignment='top', bbox=dict(facecolor='white', alpha=0.8))

fig.tight_layout()
plt.show()
```
